# End-to-End HR Analytics: Employee Attrition Analysis and Predictive Insights



```python
#import pandas for Analysis 
import pandas as pd
```

```python
#Adding csv file
df=pd.read_csv(r"C:\Users\rguun\OneDrive\Documents\HR-.csv")
```

```python
df.head()
```

```python
#Getting the info od=f data
df.info()
```

```python
#Sum of Null values
df.isnull().sum()
```

```python
#Convert Attrition Yes/No → 1/0
df["Attrition"]=df["Attrition"].apply(lambda x: 1 if x=="Yes" else 0)
```

```python
df["Attrition"].value_counts()
```

```python
df.head()
```

```python
#Creating Age Group column
df["Age_gorup"]=pd.cut(
    df["Age"],
    bins=[17,25,35,45,60],
    labels=["18-25","26-35","36-45","46+"]
)
```

```python
df.head()
```

```python
df["MonthlyIncome"].describe()
```

```python
#Creating Income Band column
df["Income_Band"]=pd.cut(
    df["MonthlyIncome"],
    bins=[0,5000,10000,df["MonthlyIncome"].max()],
    labels=["Low","Medium","High"]
)
```

```python
df["Income_Band"].head()
```

```python
df["YearsAtCompany"].max()
```

```python
#Extreme value, not outlier
df[df["YearsAtCompany"] == 40][["Age", "YearsAtCompany"]]
```

```python

# Creating Tenure Group (0–3, 4–7, 8+ years)
df["Tenure_group"]=pd.cut(
    df["YearsAtCompany"],
    bins=[-1,3,7,df["YearsAtCompany"].max()],
    labels=["0-1 Years","4-7 Years","8+ Years"]
)
```

```python
#Total employees
total_employees=df.shape[0]
print(total_employees)
```

```python
#Attrition count
#1->Yes and 0->No

df["Attrition"].value_counts()
```

```python
#Attrition rate
round(df["Attrition"].mean()*100,0)
```

```python
#Average income (left vs stayed)
df.groupby("Attrition")["MonthlyIncome"].mean().round(2)
```

```python
#Attrition by gender
round(df.groupby("Gender")["Attrition"].mean()*100,2)
```

```python
#Attrition by department
round(df.groupby("Department")["Attrition"].mean()*100,2)
```

```python
#Attrition by job role
round(df.groupby("JobRole")["Attrition"].mean()*100,2)
```

```python
#Overtime vs attrition
round(df.groupby("OverTime")["Attrition"].mean()*100,2)
```

```python
#Job satisfaction vs attrition
#1--> Low and 4-->high
round(df.groupby("JobSatisfaction")["Attrition"].mean()*100,2)
```

```python
#Years at company vs attrition
round(df.groupby("Tenure_group")["Attrition"].mean()*100,2)
```
